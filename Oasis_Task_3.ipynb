```python
# Now you can convert the data types
x_train = x_train.astype(int)
y_train = y_train.astype(int)
x_test = x_test.astype(int)
y_test = y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler # Changed standardscalar to StandardScaler
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression # Changed LinearRegrssion to LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
